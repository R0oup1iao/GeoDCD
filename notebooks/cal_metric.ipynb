```python
import pandas as pd 
import numpy as np
import json
from pathlib import Path

results = {}
for dataset_path in Path("../results").iterdir():
    dataset_name = dataset_path.stem
    all_results = []
    for exp_path in dataset_path.iterdir():
        exp_name = exp_path.stem
        
        with open(exp_path / 'metrics.json', "r") as f:
            result_data = json.load(f)
            all_results.append(result_data)
        
    results[dataset_name]=pd.DataFrame(all_results)
```

```python
for dataset_name, df in results.items():
    print(f"Dataset: {dataset_name}")
    mean_results = df.mean(numeric_only=True)
    std_results = df.std(numeric_only=True)
    
    for metric in mean_results.index:
        mean_val = mean_results[metric]
        std_val = std_results[metric]
        print(f"  {metric}: {mean_val:.4f} ± {std_val:.4f}")
    print()
```
