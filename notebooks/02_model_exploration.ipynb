```python
import numpy as np
import pandas as pd
from pathlib import Path

def sparse_to_dense(df):
    src = df.iloc[:, 0].values
    tgt = df.iloc[:, 1].values
    n_size = max(src.max(), tgt.max()) + 1
    adj_matrix = np.zeros((n_size, n_size), dtype=int)
    adj_matrix[src, tgt] = 1
    
    return adj_matrix

data_path = Path(r"../data/real/raw_data/fMRI")

for i in range(1, 29):
    gt_file = data_path / f"sim{i}_gt_processed.csv"
    ts_file = data_path / f"timeseries{i}.csv"
    
    gt_df = pd.read_csv(gt_file,header=None)
    ts_df = pd.read_csv(ts_file)
    
    gt_np = sparse_to_dense(gt_df)
    ts_np = ts_df.values
    
    np.save(rf"../data/real/fMRI/gt_{i-1}", gt_np)
    np.save(rf"../data/real/fMRI/data_{i-1}", ts_np)
    print(f"Saved {i-1}, ts shape: {ts_np.shape}, gt shape: {gt_np.shape}")
    
    
    
```
